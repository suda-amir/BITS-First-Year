```python

# Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
```

```python

"""
    A simple neural network written in Keras (TensorFlow backend) to classify the IRIS data
"""

iris_data = load_iris() # load the iris dataset

print('Example data: ')
print(iris_data.data[:5])
print('Example labels: ')
print(iris_data.target[:5])

x = iris_data.data
y_ = iris_data.target.reshape(-1, 1) # Convert data to a single column

# One Hot encode the class labels
encoder = OneHotEncoder(sparse=False)
y = encoder.fit_transform(y_)
#print(y)

# Split the data for training and testing
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.20)

# Build the model

model = Sequential()

#input layer
model.add(Dense(10, input_shape=(4,), activation='softmax', name='fc1'))
#hidden layer
model.add(Dense(5, activation='softmax', name='fc2'))
#output layer
model.add(Dense(3, activation='softmax', name='output'))

# Adam optimizer with learning rate of 0.001
optimizer = SGD(lr=0.0001)
model.compile(optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

print('Neural Network Model Summary: ')
print(model.summary())

# Train the model
model.fit(train_x, train_y, verbose=2, batch_size=5, epochs=1000)

# Test on unseen data

results = model.evaluate(test_x, test_y)

print('Final test set loss: {:4f}'.format(results[0]))
print('Final test set accuracy: {:4f}'.format(results[1]))
```

```python
#import Dataset


iris = load_iris()



X = iris.data
y = iris.target.reshape(-1, 1) # Convert data to a single column

# One Hot encode the class labels
encoder = OneHotEncoder(sparse=False)
y = encoder.fit_transform(y)
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from tensorflow.keras.optimizers import SGD

 # SGD optimizer with learning rate of 0.001
opt=SGD(lr=0.0001)

def nueralnetwork(unit):
    
# Build the model

    model = Sequential()

    #input layer
    model.add(Dense(unit, input_shape=(4,), activation='relu', name='fc1'))
    #hidden layer
    model.add(Dense(unit, activation='relu', name='hidden'))
    #output layer
    # softmax is the generic form of logistic activation function
    model.add(Dense(3, activation='softmax', name='output'))
 
    model.compile(opt, loss='binary_crossentropy', metrics=['accuracy'])

    print('Neural Network Model Summary: ')
    print(model.summary())

    # Train the model
    history=model.fit(X_train,y_train, verbose=2, batch_size=5, epochs=10000)

    # Test on unseen data

    results = model.evaluate(X_train,y_train)

    print('Final test set loss: {:4f}'.format(results[0]))
    print('Final test set accuracy: {:4f}'.format(results[1]))

    print(results)
    return results,history


```

```python
accuracy=[]
loss=[]
histry=[]
timer=[]
import time

for i in [5,10,15,20,25]:#
    start_time=time.time()

    result,history= nueralnetwork(i)
    time_taken=time.time() - start_time
    print(result,history.history.keys())
    accuracy.append(result[1])
    loss.append(result[0])
    histry.append(history)
    timer.append(time_taken)
    
    
    
```

```python
print(accuracy,loss)
acc=[]
lss=[]
for i in histry:
    for j in i.history['acc']:
        acc.append(j)
    for k in i.history['loss']:
        lss.append(k)
```

```python
def loss_curve():
   
    plt.plot(lss)
    plt.title('model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['loss'], loc='upper left')
    plt.show()

loss_curve()
```

```python
def acc_curve():
    plt.plot(acc)  
    plt.title('model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['accuracy'], loc='upper left')
    plt.show()

acc_curve()
```

```python
def plot_time():
    plt.plot(timer)
    plt.title("Time Taken")
    plt.xlabel('units')
    plt.ylabel('seconds')
    plt.show()
plot_time()
```
